```python
# 복합 대입 연산자 += 를 사용하여 3의 배수 처음 네 개 구하기

# 첫 번째 3의 배수
num = 3
print(f"첫 번째 3의 배수: {num}")

# 두 번째 3의 배수
num += 3
print(f"두 번째 3의 배수: {num}")

# 세 번째 3의 배수
num += 3
print(f"세 번째 3의 배수: {num}")

# 네 번째 3의 배수
num += 3
print(f"네 번째 3의 배수: {num}")
```
