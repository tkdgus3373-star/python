```python
def height_cm():
  return float(input(""))

def weight_kg():
  return float(input(""))
```

```python
# 키와 몸무게 입력 받기
height_cm = height_cm()
W = weight_kg()

# 키를 미터로 변환
H = height_cm / 100

# BMI 계산
bmi = W / (H ** 2)

# 소수 셋째 자리에서 반올림하여 출력
print(round(bmi, 2))
```

```python
def read_input():
  a, b = map(int, input().split())
  return a, b

def print_output(result):
  print(result)

a, b = read_input()
result = a + b
print_output(result)
```

```python
scores = [95, 87, 92, 88]
print("성적 순위:")
for rank, score in enumerate(scores, start=1):
    print(f" {rank}등: {score}점")
```

```python
students = ["철수", "영희", "민수"]
scores = [85, 92, 78]
grades = ["B", "A", "C"]
print("\n학생 성적표:")
for name, score, grade in zip(students, scores, grades):
    print(f" {name}: {score}점 ({grade}학점)")
```

```python
celsius = [0, 10, 20, 30, 40]
fahrenheit = list(map(lambda c: c * 9/5 + 32, celsius))
print(f"섭씨: {celsius}")
print(f"화씨: {fahrenheit}")
```

```python
numbers = [-5, 3, -1, 8, 0, -3, 7]
positive = list(filter(lambda x: x > 0, numbers))
print(f"전체 숫자: {numbers}")
print(f"양수만: {positive}")
```
