### Pandas 주요 함수 및 메서드 요약

Pandas는 두 가지 핵심 데이터 구조인 `Series`와 `DataFrame`을 제공합니다. `Series`는 1차원 배열과 유사한 객체이며, `DataFrame`은 여러 `Series`로 구성된 2차원 테이블 형식의 데이터 구조입니다.

#### 1. 데이터 생성 및 불러오기
- `pd.DataFrame()`: 딕셔너리, 리스트 등으로 DataFrame 생성
- `pd.Series()`: Series 생성
- `pd.read_csv('경로/파일명.csv')`: CSV 파일 불러오기
- `pd.read_excel('경로/파일명.xlsx')`: Excel 파일 불러오기

#### 2. 데이터 확인 및 정보 탐색
- `.head(n)`: DataFrame의 상위 `n`개 행 출력 (기본값 5)
- `.tail(n)`: DataFrame의 하위 `n`개 행 출력 (기본값 5)
- `.info()`: DataFrame의 요약 정보 (컬럼, 데이터 타입, 결측치 등) 출력
- `.describe()`: 숫자형 컬럼의 통계 요약 (평균, 표준편차, 최소/최대값 등) 출력
- `.shape`: DataFrame의 행과 열 개수 (튜플) 출력
- `.columns`: 컬럼 이름 목록 출력
- `.index`: 인덱스 목록 출력
- `.dtypes`: 각 컬럼의 데이터 타입 출력
- `.value_counts()`: Series의 고유 값과 그 빈도 출력

#### 3. 데이터 선택 및 필터링
- `df['컬럼명']`: 단일 컬럼 선택 (Series 반환)
- `df[['컬럼1', '컬럼2']]`: 여러 컬럼 선택 (DataFrame 반환)
- `df.loc[행_라벨, 열_라벨]`: 라벨 기반으로 데이터 선택
- `df.iloc[행_위치, 열_위치]`: 정수 위치 기반으로 데이터 선택
- `df[df['컬럼명'] > 값]`: 조건에 따라 행 필터링

#### 4. 데이터 조작 및 변환
- `.dropna()`: 결측치(NaN)가 있는 행 또는 열 제거
- `.fillna(값)`: 결측치를 특정 `값`으로 채우기
- `.astype(데이터타입)`: 컬럼의 데이터 타입 변경
- `.rename(columns={'이전이름': '새이름'})`: 컬럼 이름 변경
- `.drop(['컬럼명'], axis=1)`: 특정 컬럼 제거
- `.drop_duplicates()`: 중복된 행 제거
- `.sort_values(by='컬럼명', ascending=True)`: 특정 컬럼 기준으로 정렬
- `.groupby('컬럼명').agg({'다른컬럼': '함수'})`: 특정 컬럼을 기준으로 그룹화 및 집계
- `.apply(함수)`: DataFrame 또는 Series의 각 요소 또는 행/열에 함수 적용
- `.merge(다른df, on='공통컬럼')`: 두 DataFrame 병합
- `pd.concat([df1, df2])`: 두 DataFrame 연결

#### 5. 통계 및 계산
- `.sum()`: 합계
- `.mean()`: 평균
- `.median()`: 중앙값
- `.std()`: 표준편차
- `.min()`: 최소값
- `.max()`: 최대값
- `.count()`: 결측치를 제외한 개수

#### 6. 데이터 내보내기
- `.to_csv('경로/파일명.csv', index=False)`: DataFrame을 CSV 파일로 저장
- `.to_excel('경로/파일명.xlsx', index=False)`: DataFrame을 Excel 파일로 저장

```python
# 예시 코드
import pandas as pd

# 1. 데이터 생성
student_data = {
    'Name': ['Kim', 'Lee', 'Park', 'Choi'],
    'Test': [85, 92, 78, 88],
    'Assign1': [88, 95, 80, 90],
    'Assign2': [91, 89, 82, 85]
}
df_students = pd.DataFrame(student_data)
display(df_students)

# 2. 데이터 확인
print("\n--- df_students.info() ---")
df_students.info()

print("\n--- df_students.describe() ---")
display(df_students.describe())

# 3. 데이터 선택 및 필터링
print("\n--- 'Name' 컬럼 선택 ---")
display(df_students['Name'])

print("\n--- Test 점수가 80점 이상인 학생 필터링 ---")
display(df_students[df_students['Test'] >= 80])

# 4. 데이터 조작 및 변환
# 'Total' 컬럼 추가
df_students['Total'] = df_students['Test'] + df_students['Assign1'] + df_students['Assign2']
print("\n--- 'Total' 컬럼 추가 후 ---")
display(df_students)

# 'Name'으로 그룹화하여 'Total'의 평균 계산 (이 경우 Name은 고유하므로 의미가 제한적)
print("\n--- 'Name'으로 그룹화 후 'Total' 평균 ---")
display(df_students.groupby('Name')['Total'].mean())

# 5. 통계 계산
print("\n--- Test 점수의 평균 ---")
print(f"Test 평균: {df_students['Test'].mean():.2f}")

print("\n--- Total 점수의 최댓값 ---")
print(f"Total 최댓값: {df_students['Total'].max()}")

# 6. 데이터 정렬
print("\n--- 'Total' 점수 기준으로 내림차순 정렬 ---")
display(df_students.sort_values(by='Total', ascending=False))
```
