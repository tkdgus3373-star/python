```python
# 1단계 (데이터 준비)
dynasties = [
    "태조", "정종", "태종", "세종", "문종", "단종", "세조", "예종", "성종", "연산군",
    "중종", "인종", "명종", "선조", "광해군", "인조", "효종", "현종", "숙종", "경종",
    "영조", "정조", "순조", "헌종", "철종", "고종", "순종"
]

# 4단계 (기능 통합) - 사용자 입력 처리
user_input = input("순번(숫자) 또는 묘호(문자열)를 입력하세요: ")

if user_input.isdigit():
    # 2단계 (순번 → 묘호)
    num = int(user_input)
    if 1 <= num <= len(dynasties):
        print(f"{num}번째 왕의 묘호는 {dynasties[num - 1]}입니다.")
    else:
        print("유효한 순번을 입력해주세요 (1부터 27까지).")
else:
    # 3단계 (묘호 → 순번)
    king_name = user_input
    if king_name in dynasties:
        index = dynasties.index(king_name)
        print(f"{king_name}은(는) {index + 1}번째 왕입니다.")
    else:
        print(f"'{king_name}'에 해당하는 묘호를 찾을 수 없습니다.")
```

```python
score={'홍길동':100,'이순신':90,'강감찬':80}
```

```python
# 총점 계산
total_score = sum(score.values())
print(f"총점: {total_score}")

# 평균 계산
average_score = total_score / len(score)
print(f"평균: {average_score}")

# 가장 높은 점수를 받은 학생 찾기
highest_score = max(score.values())
for name, s in score.items():
    if s == highest_score:
        print(f"가장 높은 점수를 받은 학생: {name} ({highest_score}점)")
```
